### Data pre-processing

```python
import pandas as pd
gender_classification = pd.read_csv('gender-classification.csv')

# rename the columns
gender_classification = gender_classification.set_axis(['user_name', 'date', 'tweet_ID', 'Text'], axis=1, inplace=False)

# delete columns with dates and tweet id 
gender_classification = gender_classification.drop('date', 1)
gender_classification = gender_classification.drop('tweet_ID', 1)

# remove retweets
gender_classification = gender_classification[~gender_classification.Text.str.startswith('RT')]

# exclude users with less than 100 original tweets
gender_classification = gender_classification.groupby('user_name').filter(lambda x : len(x)>99)

# concatenate random sample of 100 Tweets per user
gender_classification = gender_classification.groupby('user_name').head(100).reset_index(drop=True)
gender_classification = gender_classification.groupby(['user_name'], as_index = False).agg({'Text': list})

# attach binary gender label to dataframe (i.e. concatinate original and scraped dataframe)
gender_classification = pd.merge(gender_classification, gndr_class, on='user_name', how='left')

# reset the index
gender_classification.reset_index(inplace=True)
gender_classification = gender_classification.drop(['index'], axis=1)

# lowercasing
for index in range(0, len(gender_classification)):
    gender_classification['Text'][index] = ''.join(gender_classification['Text'][index]).lower()
    
# replace all URLs with 'URL'
import re
for index in range(0, len(gender_classification)):
    gender_classification['Text'][index] = re.sub(r'http\S+', 'URL', ''.join(gender_classification['Text'][index]))

# replace all usernames with 'USER'
for index in range(0, len(gender_classification)):
    gender_classification['Text'][index] = re.sub(r'\B@\w+', 'USER', ''.join(gender_classification['Text'][index]))

# create an additional tokenized text column (word tokenization)
import numpy as np
from nltk.tokenize import word_tokenize

gender_classification['tokenized_text'] = np.nan
for index in range(0, len(gender_classification)):
    gender_classification['tokenized_text'][index] = word_tokenize(gender_classification['Text'][index])
```

### Feature extraction

```python
# number of words
gender_classification['n_words'] = gender_classification['tokenized_text'].str.len()

# number of characters
gender_classification['n_char'] = gender_classification['Text'].str.len()

# average lenght of words (n_char divided by n_words)
gender_classification['avg_len'] = gender_classification['n_char']/gender_classification['n_words']

# average number of words per tweet
gender_classification['avg_tweet_len'] = gender_classification['n_words']/100

# vocabulary diversity (number of tokens divided by number of types) higher is more divers
gender_classification['divers'] = np.nan
for index in range(0, len(gender_classification)):
    gender_classification['divers'][index] = len(set(gender_classification['tokenized_text'][index])) / gender_classification['n_words'][index]

# overall polarity
from textblob import TextBlob
gender_classification['polarity'] = np.nan

for index in range(0, len(gender_classification)):
    text_blob = TextBlob(gender_classification['Text'][index])
    gender_classification['polarity'][index] = text_blob.sentiment.polarity
    
# overall subjectivity
gender_classification['subjectivity'] = np.nan

for index in range(0, len(gender_classification)):
    text_blob = TextBlob(gender_classification['Text'][index])
    gender_classification['subjectivity'][index] = text_blob.sentiment.subjectivity

# PoS-tagging (raw - after data splitting we will convert them to proportions and determine the cutoff)
import nltk
gender_classification['PoS'] = np.nan

for index in range(0, len(gender_classification)):
    gender_classification['PoS'][index] = nltk.pos_tag(gender_classification['tokenized_text'][index])

display(gender_classification)
```

### Training and development (i.e. hyperparameter selection)

#### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# hyperparameters to consider
n_estimators = [70, 100, 130, 160] 
max_features = [200, 500, 1000, 2000] 
max_depth = [10, 20, 40, 60] 

best_acc = 0
best_estimate = 0
best_feature = 0
best_depth = 0

for estimate in n_estimators:
    for feature in max_features:
        for depth in max_depth:
            
            RF = RandomForestClassifier(criterion = 'gini',
                                        n_estimators = estimate,
                                        max_features = feature,
                                        max_depth = depth,
                                        random_state = 0)
            RF.fit(x_train, y_train)
            y_pred = RF.predict(x_devlp)
            
            print('Hyperparameter setting: n_estimators =', estimate, 'max_features =', feature, 'max_depth =', depth)
            print(accuracy_score(y_devlp, y_pred))
            print(confusion_matrix(y_devlp, y_pred))
            print(classification_report(y_devlp, y_pred))
            
            acc = accuracy_score(y_devlp, y_pred)
            if acc > best_acc:
                best_acc = acc
                best_estimate = estimate
                best_feature = feature
                best_depth = depth
```

```python
print('The best RF model has an accuracy of:', best_acc)
print('Hyperparameter setting: n_estimators =', best_estimate, 'max_features =', best_feature, 'max_depth =', best_depth)
```

#### Support Vector Machine

```python
from sklearn import svm

# hyperparameters to consider
setting_c = [0.001, 0.01, 0.1, 1]
setting_kernel = ['linear', 'rbf', 'sigmoid']
setting_gamma = [0.05, 0.1, 0.5]

best_acc_svm = 0
best_c = 0
best_kernel = 0
best_gamma = 0

for kernel in setting_kernel:
    for c in setting_c:
        for gamma in setting_gamma:
            
            SVM = svm.SVC(kernel = kernel,
                         C = c,
                         gamma = gamma,
                         random_state = 0) 
            SVM.fit(x_train, y_train)
            y_pred = SVM.predict(x_devlp)
            
            print('Hyperparameter setting: kernel =', kernel, 'C =', c, 'gamma =', gamma)
            print(accuracy_score(y_devlp, y_pred))
            print(confusion_matrix(y_devlp, y_pred))
            print(classification_report(y_devlp, y_pred))
            
            acc_svm = accuracy_score(y_devlp, y_pred)
            if acc_svm > best_acc_svm:
                best_acc_svm = acc_svm
                best_c = c
                best_kernel = kernel
                best_gamma = gamma
```
